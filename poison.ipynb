# *PREPROCESSING*

```python
df.astype('object').columns
```

```python
dummies=pd.get_dummies(df[['class','cap-diameter', 'cap-shape', 'cap-surface', 'cap-color',
       'does-bruise-or-bleed', 'gill-attachment', 'gill-spacing', 'gill-color',
       'stem-height', 'stem-width', 'stem-root', 'stem-surface', 'stem-color',
       'veil-type', 'veil-color','has-ring','ring-type','spore-print-color',
       'habitat', 'season']],drop_first=True)
```

```python
df=df.drop(['class','cap-diameter', 'cap-shape', 'cap-surface', 'cap-color',
       'does-bruise-or-bleed', 'gill-attachment', 'gill-spacing', 'gill-color',
       'stem-height', 'stem-width', 'stem-root', 'stem-surface', 'stem-color',
       'veil-type', 'veil-color','has-ring','ring-type','spore-print-color',
       'habitat', 'season'],axis=1)
```

```python
df=pd.concat([df,dummies],axis=1)
```

# *CREATING THE MODEL*

```python
#splitting the data and evaluating the model
```

```python
X=df.drop('class_p',axis=1).values

y=df['class_p'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X)
X_test=scaler.fit(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rm_model=RandomForestClassifier()
```

```python
lr_model=LogisticRegression()
svm_model=SVC(C=1,kernel='rbf')
```

```python
X_train.reshape(-1, 1)
```

```python
lr_model.fit(X_train,y_train)
```

```python
svm_model.fit(X_train,y_train)
```

```python
rm_model.fit(X_train,y_train)
```

```python
preds=lr_model.predict(X_test)
svm_preds=svm_model.predict(X_test)
```

```python
rm_preds=rm_model.predict(X_test)
```

# *EVALUATION OF THE MODEL*

```python
print(f'LogisticReg:{lr_model.score(X_test,y_test)}')
print(f'SVM:{svm_model.score(X_test,y_test)}')
```

```python
print(f'RandomForrest:{rm_model.score(X_test,y_test)}')  #100% accuracy! this model is doing fantastic!
```
